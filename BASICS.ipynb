How to define function?
1. def function_name:
2.   Business logic

```python
def calc(x,y): #To make your code concise
  """ This function do the addition of two no \n you need to provide both x and y as int"""
  print("Addition of two no is",x+y)
  print("Sub of two no is",x-y)
  print("Mul of two no is",x*y)
  print("Div of two no is",x/y)
```

Returning a multiple value

```python
def calc(x,y): #To make your code concise
  """ This function do the addition of two no \n you need to provide both x and y as int \n and it will return you those value"""
  add=x+y
  sub=x-y
  mul=x*y
  div=x/y
  return add,sub,mul,div
```

```python
import pandas as pd
df=pd.read_csv('USA_Housing.csv')
df.head()
```

```python
df.info()
```

```python
df['Address'].nunique()
```

```python
df.drop('Address',axis=1,inplace=True)
```

```python
df.info()
```

```python
X=df.drop('Price',axis=1)
```

```python
X
```

```python
y=df['Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain ,ytest=train_test_split(X,y,test_size=0.10)
```

```python
Xtrain
```

```python
ytrain
```

```python
Xtest
```

```python
ytest
```

**Model Selection**

```python
from sklearn.linear_model import LinearRegression
mdl=LinearRegression()
mdl.fit(Xtrain,ytrain)
```

```python
Pred=mdl.predict(Xtest)
```

```python
import matplotlib.pyplot as plt
plt.scatter(ytest,Pred)
```

```python
mdl.score(Xtest,ytest)
```
